```python
import tensorflow as tf
```

```python
a = tf.constant([1, 2])
b = tf.constant([3, 4])
```

```python
c = tf.add(a, b)
c.numpy()
```

```python
d = tf.subtract(a, b)
d.numpy()
```

```python
a_1 = tf.constant([[1, 2], [3, 4]])
b_1 = tf.constant([[2, 3], [4, 5]])
```

```python
m = tf.multiply(a_1, b_1)
m.numpy()
```

```python
mm = tf.matmul(a_1, b_1)
mm.numpy()
```

```python
d = tf.divide(a, b)
d.numpy()
```

```python
a = tf.constant([1,2,3,4,5,6])
a.shape
```

```python
a = tf.reshape(a, shape=(2,3))
a.shape
```

```python
b = tf.expand_dims(a, -1)
b.shape
```

```python
b = tf.constant([[1,2],[3,4]])
```

```python
b = tf.square(b)
b
```

```python
from numpy import float32
b = tf.sqrt(tf.cast(b, dtype=float32))
b
```

```python
min = tf.reduce_min(b, axis=None)
min.numpy()
```

```python
min = tf.reduce_min(b, axis=0)
min
```

```python
max = tf.reduce_max(b, axis=1)
max
```
